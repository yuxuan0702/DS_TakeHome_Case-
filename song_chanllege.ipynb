# Create metrics represent user engagment. Display top 3 and bottom 3 states in terms of user engagement 

`User Engagement Metrics`: 
Average plays per user in the given state.


```python
# define user number and play number 
user_number = df1.groupby('user_state')[['user_id']].count().rename(columns = {'user_id':'user_number'})
play_number = df.groupby('user_state')[['user_id']].count().rename(columns = {'user_id': 'play_number'})
```

```python
# define user engagment 
df3 = pd.merge(user_number, play_number, left_index=True, right_index=True)
df3['user_engagement'] = df3.play_number/df3['user_number']
df3.sort_values('user_engagement',ascending = False,inplace = True)
```

# Build a function that takes as an input any of the songs in the data and returns the most likely song to be listened next. 

`Similarity` : Number of users who listen them together

```python
# step 1 build the song - user matrix
song_user = df.pivot_table(index='song_played',
               columns='user_id',
               values='id',
               fill_value=0,
               aggfunc='count')

song_user = (song_user>0).astype(int)
```

```python
song_user
```

```python
# step 2 build song-song similarity matrix
song_user_norm = normalize(song_user, axis=1)
similarity = np.dot(song_user_norm, song_user_norm.T)
similarity_df = pd.DataFrame(similarity,
                             index=song_user.index,
                             columns=song_user.index)

similarity_df
```

```python
# Step 3: find the top-k most similar songs
def find_topk(song,k = 1 ):
    df = similarity_df.loc[song].sort_values(ascending=False)[1:k + 1].reset_index()
    df = df.rename(columns={'song_played': 'Song', song: 'Similarity'})
    return df
    
```

```python
find_topk('Yellow Submarine',10)
```
